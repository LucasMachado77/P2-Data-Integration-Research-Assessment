```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
chromedriver_path = r'C:\chromedriver\chromedriver.exe'
service = Service(executable_path=chromedriver_path)
driver = webdriver.Chrome(service=service)

driver.get("https://portal2.ipt.pt/pt/quem_e_quem/")
time.sleep(2)

resultados = []

NUM_PAGINAS = 26

for pagina in range(NUM_PAGINAS):
    print(f"Extraindo página {pagina + 1}/{NUM_PAGINAS}")
    time.sleep(1.2)  # Garante carregamento

    perfis = driver.find_elements(By.CSS_SELECTOR, 'div#perfil')
    for perfil in perfis:
        texto = perfil.text.strip()
        linhas = texto.split('\n')
        nome = linhas[0] if linhas else ''
        email = ''
        membro_de = []
        leciona_em = []
        cargos = []

        for l in linhas[1:]:
            l = l.strip()
            if l.startswith('Email'):
                email = l.replace('Email', '').strip()
            elif l.startswith('Membro de'):
                membro_de.append(l.replace('Membro de', '').strip())
            elif l.startswith('Leciona em'):
                leciona_em.append(l.replace('Leciona em', '').strip())
            elif (l.startswith('Coordenador') or l.startswith('Diretor') or l.startswith('Responsável')):
                cargos.append(l.strip())
        resultados.append({
            'Nome': nome,
            'Email': email,
            'Membro de': '; '.join(membro_de),
            'Leciona em': '; '.join(leciona_em),
            'Cargos': '; '.join(cargos),
            'Raw': texto
        })

    # Clica no botão "Seguinte", exceto na última
    if pagina < NUM_PAGINAS - 1:
        try:
            li_seguinte = driver.find_element(By.XPATH, "//li[normalize-space(text())='Seguinte']")
            try:
                li_seguinte.click()
            except Exception:
                driver.execute_script("arguments[0].click();", li_seguinte)
        except Exception as e:
            print("Erro ao encontrar/clicar Seguinte:", e)
            break

driver.quit()

df_result = pd.DataFrame(resultados)
df_result.to_excel("quem_e_quem_completo.xlsx", index=False)
print("Finalizado! Arquivo quem_e_quem_completo.xlsx gerado.")
```
